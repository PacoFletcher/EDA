```python
barrio = rentas["Distrito"].unique()
barrio
```

```python
dfs = []
for b in barrio:
    if b == "BILBAO" or b == "Donostia / San Sebastián" or b == "VITORIA-GASTEIZ":
        temp = rentas.loc[rentas["Distrito"] == b].drop(columns="Distrito",axis = 1).set_index("Renta").T.reset_index().rename(columns={"index":"Año"}).copy()
        temp["Distrito"] = b
        dfs.append(temp)
capitales = pd.concat(dfs, axis = 0, ignore_index=True)
capitales
```

```python
capitales["Año"] = pd.to_datetime(capitales["Año"],format = "%Y")
capitales.rename(columns={"Distrito":"Ciudad"},inplace=True)
```

```python
capitales.dtypes
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(data = capitales, x = "Año", y = "Renta total", hue = "Ciudad")
plt.title("Renta total capitales provincia")
plt.xticks(rotation = "vertical");
```

```python
triang = capitales.iloc[13:,1:-1].corr()
mascara = np.triu(np.ones_like(triang,dtype=bool),1)
mascara
```

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(capitales.iloc[:,1:-1].corr(),
            mask = mascara,
            annot=True,
            linewidths=.5,
            ax=ax,
            vmin=0,
            vmax=1,
            cmap = "Reds")
plt.title("Tipos de Renta en las capitales de provincia");
```
