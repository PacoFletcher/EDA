```python
plt.figure(figsize=(10,9));
sns.lineplot(data = capitales, x = "Año", y = "Renta total", hue = "Ciudad")
sns.lineplot(data = capitales, x = "Año", y = "Renta del trabajo", hue = "Ciudad")
plt.xticks(rotation = "vertical");
```

```python
triang = capitales.iloc[13:,1:-1].corr()
mascara = np.triu(np.ones_like(triang,dtype=bool),1)
mascara
```

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(capitales.iloc[:,1:-1].corr(),
            mask = mascara,
            annot=True,
            linewidths=.5,
            ax=ax,
            vmin=0,
            vmax=1,
            cmap = "Reds")
plt.title("Tipos de Renta en las capitales de provincia");
```
