```python
lista_top = ["Abando", "Indautxu", "Uribarri", "Casco", "Basurto", "Rekalde", "Otxarkoaga"]
```

```python
variacion = prueba[prueba["Distrito"].isin(lista_top)]
variacion
```

```python
variacion.index[0]
```

```python
diferencia = variacion[variacion.index.year.isin(range(2011,2022))]
diferencia
```

```python
var_abs = []
var_porc = []
for i in diferencia["Distrito"].unique():
    print(i)
    x = diferencia[diferencia["Distrito"] == i]["Precio m2"].iloc[-1]
    y = diferencia[diferencia["Distrito"] == i]["Precio m2"].iloc[0]
    var_abs.append(round((x-y),2))
    var_porc.append(round((x-y)/y*100,2))
print(var_abs)
print(var_porc)
```

```python
calc = pd.DataFrame({"Distrito": lista_top[:-1],
                     "Variación absoluta": var_abs,
                     "Variación porcentual": var_porc})
calc
```
