```python
alquileres = alq_bilbo
dfs = [alq_abando,alq_basurto,alq_begonia,alq_casco,alq_deusto,alq_ibaiondo,alq_indautxu,alq_rekalde,alq_uribarri,]
```

```python
for i in dfs:
    alquileres = pd.concat([alquileres,i],axis=0,ignore_index=True)
alquileres
```

```python
alquileres.columns
```

```python
alquileres["Precio m2"] = alquileres["Precio m2"].str.replace(" €/m2","")
#alquileres["Precio m2"] = np.float64(alquileres["Precio m2"])
alquileres
```

```python
prueba = alquileres[alquileres["Precio m2"] != "n.d."]
```

```python
prueba
```

```python
prueba.loc[:,"Precio m2"] = np.float64(prueba["Precio m2"])
```

```python
prueba.loc[:,"Mes"] = pd.to_datetime(prueba["Mes"],dayfirst=True)
prueba
```

```python
alq = prueba 
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data = prueba[prueba["Distrito"] != "BILBAO"], x = "Mes", y = "Precio m2", hue = "Distrito")
sns.lineplot(data = prueba[prueba["Distrito"] == "BILBAO"], x = "Mes", y = "Precio m2",linewidth = 5, label = "BILBAO")
plt.title("Evolución precio de alquiler en Bilbao €/m2");
```

```python
alq[alq == "n.d."]
```

```python
prueba2 = alq.drop(alq[alq["Variación anual"] == "n.d."].index)
prueba2
```

```python
prueba2["Variación anual"] = prueba2["Variación anual"].str.replace(",",".")
prueba2["Variación anual"] = prueba2["Variación anual"].str.replace("%","")
```

```python
prueba2["Variación anual"] = np.float64(prueba2["Variación anual"])
prueba2
```

```python
ipc = pd.read_csv("Data/ipc_euskadi.csv", sep = ";", encoding="Windows 1250")
ipc["Total"] = ipc["Total"].str.replace(",",".")
ipc["Total"] = np.float64(ipc["Total"])
ipc
```

```python
ipc["Periodo"] = pd.to_datetime(ipc["Periodo"],format= "%Y")
```

```python
x = np.array(ipc["Periodo"])
x[2:14]
```

```python
plt.figure(figsize=(10,9))
sns.lineplot(data = prueba2[prueba2["Distrito"] == "BILBAO"], x = "Mes", y = "Variación anual", label = "Alquiler m2 Bilbao")
sns.lineplot(data = ipc[ipc["Grupos ECOICOP"] == "04 Vivienda, agua, electricidad, gas y otros combustibles"], x = "Periodo", y = "Total", label = "IPC Euskadi")
plt.ylabel("Variación %")
plt.xlim(x[12],x[1])
plt.title("Variación porcentual del IPC de vivienda + suministros y precio del alquiler en Bilbao");
```
