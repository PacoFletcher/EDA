```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from dateutil import parser
```

```python
bilbo = pd.read_csv("Data/renta_bilbo.csv")
bilbo["Año"] = pd.to_datetime(bilbo["Año"])
bilbo.dtypes
```

```python
ipc = pd.read_csv("Data/ipc_ok.csv")
ipc["Periodo"] = pd.to_datetime(ipc["Periodo"])
ipc
```

```python
ipc_abs = pd.read_csv("Data/ipc_abs_ok.csv", sep = ",")
ipc_abs
```

```python
ipc_abs["Periodo"] = pd.to_datetime(ipc_abs["Periodo"])
ipc_abs.sort_values(["Grupos ECOICOP","Periodo"], ascending = [False,True], inplace=True)
ipc_abs
```

```python
ipc_abs.rename(columns={"Grupos ECOICOP":"Grupo"}, inplace=True)
```

```python
ipc_abs.columns
```

```python
val_2006_gen = ipc_abs[(ipc_abs["Periodo"].dt.year == 2006) & (ipc_abs["Grupo"] == "Índice general")]["Total"]
v_g_2006 = val_2006_gen.iloc[0]
v_g_2006
```

```python
val_2006_viv = ipc_abs[(ipc_abs["Periodo"].dt.year == 2006) & (ipc_abs["Grupo"] == "04 Vivienda, agua, electricidad, gas y otros combustibles")]["Total"]
v_v_2006 = val_2006_viv.iloc[0]
v_v_2006
```

```python
var_an = (ipc_abs[ipc_abs["Grupo"] == "Índice general"]["Total"]-v_g_2006)/v_g_2006*100
var_an
```

```python
var_viv_an = (ipc_abs[ipc_abs["Grupo"] == "04 Vivienda, agua, electricidad, gas y otros combustibles"]["Total"]-v_v_2006)/v_v_2006*100
var_viv_an
```

```python
var_an = pd.concat([var_an,var_viv_an], axis = 0)
var_2006 = {"Var_2006":var_an}
var_2006 = pd.DataFrame(var_2006)
```

```python
var_2006
```

```python
ipc_abs = pd.concat([ipc_abs,var_2006], axis = 1)
```

```python
ipc_abs
```

```python
def variacion(df,bar,col):
    print(col)
    var = [0]
    j = df[df["Distrito"] == bar][col][0]
    for i in df[df["Distrito"] == bar][col][1:]:
        var.append(round((i-j)/i*100,2))
    #var[col] = pd.Series(var)
    print(var)
    return(var)
```

```python
var_gen = variacion(bilbo,"BILBAO","Renta total")
#var_gen.append(variacion(bilbo,"BILBAO","Renta total"))
df = bilbo[bilbo["Distrito"] == "BILBAO"]
df["Var_2006"] = var_gen
df
```

```python
df["Año"].iloc[-1].year
```

```python
plt.figure(figsize=(10,9))
sns.lineplot(data = ipc_abs[ipc_abs["Grupo"] == "Índice general"], x = "Periodo", y = "Var_2006", hue = "Grupo", marker = 'o')
sns.lineplot(data = df, x = "Año", y = "Var_2006", label = "Evolución Renta Bilbao", marker='o')
plt.ylabel("Variacón porcentual respecto 2006")
plt.title("Variación porcentual del IPC y de las rentas en Bilbao respecto 2006")
plt.xlim(df["Año"].iloc[0],df["Año"].iloc[-1]);
```

```python
top5_bar_up = bilbo[(bilbo["Año"].dt.year == 2022) & (bilbo["Distrito"] != "BILBAO")][["Distrito", "Renta total"]].sort_values("Renta total",ascending=False).head(5)["Distrito"]
print(top5_bar_up)
top_up = bilbo[bilbo["Distrito"].isin(top5_bar_up)]
top_up
```

```python
def mascara(df):
    triang = df.iloc[:,1:-1].corr()
    masc = np.triu(np.ones_like(triang,dtype=bool),1)
    return(masc)
```

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(top_up.iloc[:,1:-1].corr(),
            mask = mascara(top_up),
            annot=True,
            linewidths=.5,
            ax=ax,
            vmin=0,
            vmax=1,
            cmap = "Reds")
plt.xlabel("")
plt.ylabel("")
plt.title("Tipos de Renta en las capitales de provincia");
```

```python
top5_bar_low = bilbo[(bilbo["Año"].dt.year == 2022) & (bilbo["Distrito"] != "BILBAO")][["Distrito", "Renta total"]].sort_values("Renta total").head(5)["Distrito"]
top5_bar_low
top_low = bilbo[bilbo["Distrito"].isin(top5_bar_low)]
top_low
```

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(top_low.iloc[:,1:-1].corr(),
            mask = mascara(top_low),
            annot=True,
            linewidths=.5,
            ax=ax,
            vmin=0,
            vmax=1,
            cmap = "Reds")
plt.xlabel("")
plt.ylabel("")
plt.title("Tipos de Renta en las capitales de provincia");
```

```python
plt.figure(figsize=(10,9))
sns.lineplot(data = top_up, x = "Año", y = "Renta total", hue = "Distrito")
sns.lineplot(data = top_low, x = "Año", y = "Renta total", hue = "Distrito")
sns.lineplot(data = bilbo[bilbo["Distrito"] == "BILBAO"], x = "Año", y = "Renta total", linewidth= 5, label = "BILBAO")
plt.legend(framealpha = 0, frameon = True, loc = "upper left");
```
